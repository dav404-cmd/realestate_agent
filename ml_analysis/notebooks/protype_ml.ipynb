## data preprocessing 

```python
import ast

def parse_range(x):
    if isinstance(x, str) and x.startswith("[") and x.endswith("]"):
        return ast.literal_eval(x)
    return None  # or np.nan

df["floor_clean"] = df["floor"].apply(parse_range)


df["floor_min"] = df["floor_clean"].apply(lambda x: x[0] if x else None)
df["floor_max"] = df["floor_clean"].apply(lambda x: x[1] if x else None)
```

```python
df = df.drop(columns=["floor","id","source","scraped_at","building_description","building_name","date_updated","gross_yield","unit_number","unit_summary","url","next_update_schedule"])
```
